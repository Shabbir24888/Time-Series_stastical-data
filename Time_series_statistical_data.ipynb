## Statistical data graphs

```python
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.cm as cm    ## cm stands for color maps
import matplotlib.pyplot as plt
from pylab import rcParams
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

```python
import seaborn as sb
sb.set_style('whitegrid')

address = 'E:\Jupyter notebook\Linked_in learning DS\mtcars\mtcars.csv'
cars = pd.read_csv(address)


mpg=cars['mpg']

mpg.plot(kind='hist')

```

```python
plt.hist(mpg)
plt.plot()
```

```python
sb.distplot(mpg)
```

### Scatter plot in action

```python
cars.plot(kind='scatter',x='hp',y='mpg',c=['darkgray'],s=150)
```

```python
## Scatter plot using seaborn
sb.regplot(x='hp',y='mpg',data=cars,scatter=True)
```

### Scatter plot matrix

```python
sb.pairplot(cars)
```

```python
cars_subset = cars[['mpg','disp','hp','wt']]
sb.pairplot(cars_subset)
plt.show()
```

### Box plot

```python
cars.boxplot(column='mpg',by='am')
cars.boxplot(column='wt',by='am')
```

```python
## box plot using seaborn

sb.boxplot(x='am',y='mpg',data=cars)
```
